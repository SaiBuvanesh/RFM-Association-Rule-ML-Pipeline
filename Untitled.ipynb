```python
rfm_data = rfm[['recency', 'frequency', 'monetary']]
rfm_data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_data)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(2, 11), wcss, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
rfm['cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_profile = rfm.groupby('cluster')[['recency','frequency','monetary']].mean()
cluster_profile
```

```python
rfm['segment'] = rfm['cluster'].map({
    2: 'Champions',
    3: 'Loyal Customers',
    4: 'Big Spenders (At Risk)',
    0: 'Low Value',
    1: 'Lost'
})
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

transactions = final_df['description'].tolist()

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)
basket.head()
```

```python
frequent_itemsets = apriori(
    basket,
    min_support=0.01,   # you can tune this
    use_colnames=True
)

frequent_itemsets.sort_values('support', ascending=False).head()
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.3
)

rules = rules.sort_values(['lift', 'confidence'], ascending=False)
rules.head()
```

```python
rules_clean = rules[['antecedents','consequents','support','confidence','lift']]
rules_clean.head()
```

```python
def recommend_product(product_name, rules, top_n=5):
    product_name = product_name.upper()
    
    filtered = rules[rules['antecedents'].apply(lambda x: product_name in x)]
    
    filtered = filtered.sort_values('confidence', ascending=False)
    
    return filtered[['antecedents','consequents','support','confidence','lift']].head(top_n)
```

```python
recommend_product("SET/6 RED SPOTTY PAPER CUPS", rules_clean)
```

```python
def recommend_for_basket(basket_items, rules, top_n=5):
    basket_items = [item.upper() for item in basket_items]

    # rules where antecedent is subset of basket
    matched_rules = rules[rules['antecedents'].apply(lambda x: set(x).issubset(set(basket_items)))]
    
    # sort by confidence and lift
    matched_rules = matched_rules.sort_values(['confidence', 'lift'], ascending=False)
    
    recommendations = []

    for _, row in matched_rules.iterrows():
        for item in row['consequents']:
            if item not in basket_items:
                recommendations.append(item)

    # remove duplicates, keep order
    recommendations = list(dict.fromkeys(recommendations))

    return recommendations[:top_n]
```

```python
basket = [
    "SET/6 RED SPOTTY PAPER CUPS",
    "SET/6 RED SPOTTY PAPER PLATES"
]

recommend_for_basket(basket, rules_clean)
```
